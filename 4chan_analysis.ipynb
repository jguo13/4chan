```
import pandas as pd
from datetime import datetime

df = pd.read_csv("/mg_crawled_data.csv")

#hack to bin by day
df['datetime_cut'] = ''
for index, row in df.iterrows():
    row["datetime_cut"] = row["datetime"].split(' ')[0]

    df.loc[index] = row

hist_data = df["datetime_cut"].value_counts()
x = hist_data.index
y = hist_data


zipped = list(zip(x, y))
zipped.sort(key = lambda t: t[0])
x, y = zip(*zipped)
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize=(24,12))

xlabel1 = x
x_cutlabel1 = []
for date in xlabel1:
    x_cutlabel1.append(str(date).split()[0])

ax.plot_date(x, y, 'k--', color = "mediumpurple",linewidth=4)


ax.fill_between(x, y, color='lavender')

ax.set_xticks(xlabel1)


ax.set_xticklabels(x_cutlabel1, rotation=45)

axis_font = {'fontname':'Arial', 'size':'24'}


ax.set_ylabel("Number of Posts", **axis_font)
ax.tick_params(axis='both', which='major', labelsize=14)



fig.suptitle('Posts containing u/Maxwellhill refrences', fontsize = 24)



plt.show()

```
